```python
import pandas as pd
import json
```

```python
data=pd.read_json("t20_json_files/t20_wc_match_results.json")
data.head()
```

```python
df=pd.DataFrame(data.at[0,"matchSummary"])
df.head()
```

```python
#using the scorecord id as primary id and changeing that to match id
df.rename({"scorecard":"matchid"},axis=1,inplace=True)
df.head()
```

```python
with open("t20_json_files/t20_wc_batting_summary.json") as f:
    data=json.load(f)
    all_records=[]
    for rec in data:
        all_records.extend(rec["battingSummary"])
df_batting=pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting["out/not_out"]=df_batting.dismissal.apply(lambda x: "not_out" if x==""else "out")
df_batting
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
df_batting
```

```python
df_batting["batsmanName"]=df_batting["batsmanName"].apply(lambda x: x.replace("â€",""))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
#combining the team1 and team2 of df to map with the match id in batting df
#for that..
match_dic={}
for index, row in df.iterrows():
    key1= row["team1"] + " Vs " + row["team2"]
    key2= row["team2"] + " Vs " + row["team1"]
    match_dic[key1]= row["matchid"]
    match_dic[key2]= row["matchid"]
match_dic
```

```python
df_batting["matchid"]=df_batting["match"].map(match_dic)
df_batting
```

```python
with open("t20_json_files/t20_wc_bowling_summary.json") as f:
    data=json.load(f)
    all_records=[]
    for rec in data:
        all_records.extend(rec["bowlingSummary"])
df_bowling=pd.DataFrame(all_records)
df_bowling.head()
```

```python
df_bowling["matchid"]=df_bowling["match"].map(match_dic)
df_bowling.head()
```
